# Importação de Bibliotecas

cv2: É a biblioteca OpenCV para processamento de imagens.

numpy as np: É usado para manipulação eficiente de arrays.

matplotlib.pyplot as plt: É usado para exibir as imagens.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
```

# Carregamento da Imagem
Carrega uma imagem a partir de um arquivo.

```python
# Carregando a imagem
image = cv2.imread('/content/pessoa.jpg')

# Verifica se a imagem foi carregada com sucesso
if image is None:
    print("Erro ao carregar a imagem. Verifique o caminho do arquivo.")
    exit()
```

# Translação
É o deslocamento de objetos no espaço sem alterar sua forma

```python
# Função para realizar a translação
def translate(image, x, y):
    translation_matrix = np.float32([[1, 0, x], [0, 1, y]])  # Cria uma matriz usando NumPy.
    translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0])) # Aplica a translação da imagem usando a função warpAffine do OpenCV.
    return translated_image
```

# Rotação
É uma transformação que gira objetos em torno de um ponto fixo


```python
# Função para realizar a rotação
def rotate(image, angle):
    center = (image.shape[1] // 2, image.shape[0] // 2) # Pega o centro da imagem para ser o ponto de rotação.
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0) # Cria uma matriz de rotação usando o OpenCV.
    rotated_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0])) # Aplica a transformação de rotação da imagem usando a função warpAffine do OpenCV.
    return rotated_image
```

# Escalonamento
É uma transformação que altera o tamanho de objetos

```python
# Função para realizar o escalonamento
def scale(image, scale_factor):
    scaled_image = cv2.resize(image, None, fx=scale_factor, fy=scale_factor) # Utiliza a função resize do OpenCV para aplicar o escalonamento à imagem.
    return scaled_image
```

# Funções Extra
Funções implenetadas que não estavam no escopo do projeto.

## Blur
O efeito de Blur refere-se à redução da nitidez de uma imagem

```python
# Função para aplicar desfoque (blur)
def blur(image, kernel_size):
    blurred_image = cv2.medianBlur(image, kernel_size) # Aplica o efeito de desfoque à imagem usando o método de desfoque mediano do OpenCV.
    return blurred_image
```

# Inverter Cor
Refere-se à operação de transformar as cores de uma imagem de forma que as cores claras se tornem escuras e vice-versa. Esta função aplica a operação "not" bit a bit em cada pixel da imagem, invertendo suas cores.

```python
# Função para inverter as cores da imagem
def invert_cor(image):
    inverted_image = cv2.bitwise_not(image) # Utiliza a função bitwise_not do OpenCV para inverter as cores da imagem.
    return inverted_image
```

# Ajustar contraste e brilho
Refere-se à modificação dos níveis de contraste e brilho em uma imagem para melhorar sua aparência visual.

```python
# Função para ajustar contraste e brilho
def ajust_contrast_brightness(image, alpha, beta):
    # Utiliza a função convertScaleAbs do OpenCV para ajustar contraste e brilho.
    # A fórmula aplicada é: Pixel ajustado = alpha * Pixel original + beta.
    ajusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return ajusted_image
```

# Escala de Cinza
É uma representação de imagens onde cada pixel é exibido em tons de cinza, geralmente variando de preto a branco.

```python
# Escala de Cinza
def to_grayscale(image):
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Converte a imagem para escala de cinza utilizando a função cvtColor do OpenCV.
    return cv2.cvtColor(grayscale_image, cv2.COLOR_GRAY2BGR) # Converte a imagem de volta para o formato BGR (colorido) para manter a consistência.
```

# Detecção de bordas
Refere-se à identificação dos limites ou transições entre diferentes regiões de uma imagem.



```python
# Função para detecção de borda
def edge_detection(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Converte a imagem para escala de cinza utilizando a função cvtColor do OpenCV.
    edges = cv2.Canny(gray_image, 50, 150) # Aplica o detector de bordas Canny para identificar transições de intensidade.
    edges_image = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR) # Converte a imagem de bordas de volta para o formato BGR (colorido) para visualização.
    return edges_image
```
